```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def return_pos(G): 
    return {node: (math.cos((i*360/len(G.nodes())) * 3.1415 / 180), math.sin((i*360/len(G.nodes())* 3.1415 / 180))) for i, node in enumerate(G.nodes())}
```

```python
def draw_graph(G, figsize=None):
    fig, ax = plt.subplots(figsize=figsize)
    nx.draw(G, pos=return_pos(G), with_labels=True, node_color='white')
    edge_labels = nx.get_edge_attributes(G,'weight')
    edge_labels = {tuple(keys[:2]): value for keys, value in edge_labels.items()}
    edge_labels = edge_labels if edge_labels else None
    _ = nx.draw_networkx_edge_labels(G, pos=return_pos(G),
                                 edge_labels=edge_labels)
```

```python
def draw_from_net_file(file, figsize=(15,15)):
    G = nx.read_pajek(file)
    draw_graph(G, figsize)
    
```

```python
draw_from_net_file('AnsCalT.net', figsize=(8,8))
```

```python
draw_from_net_file('stranke94.net')
```

```python
def draw_got_graph(file='got8_tyrion.csv', character='TYRION', figsize=None):
    csv_data = open('got8_tyrion.csv', "r")
    next(csv_data, None) 
    G = nx.parse_edgelist(csv_data, delimiter=',', create_using=nx.Graph(),
                          nodetype=str, data=(('weight', float),))

    fig, ax = plt.subplots(figsize=figsize)
    pos = return_pos(G)
    pos[character] = (0, 0)
    
    edge_labels = nx.get_edge_attributes(G,'weight')
    edge_labels = {tuple(keys[:2]): value for keys, value in edge_labels.items()}
    edge_labels = edge_labels if edge_labels else None
    weights = [G[u][v]['weight'] for u,v in G.edges()]
    weights = 5*np.array(weights)/max(weights)
    nx.draw(G, pos=pos, with_labels=True, width=weights, node_color='white')
    _ = nx.draw_networkx_edge_labels(G, pos=pos,
                                 edge_labels=edge_labels)
```

```python
draw_got_graph(figsize=(10,10))
```

```python
draw_from_net_file('Transatlantic.net',figsize=(8,8))
```
