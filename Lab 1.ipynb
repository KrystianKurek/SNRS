```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def draw_graph_from_am(am, positions=None):
    rows, cols = np.where(am == 1)
    edges = zip(rows.tolist(), cols.tolist())
    print(rows, cols)
    print(list(edges))
    gr = nx.Graph()
    gr.add_edges_from(edges)
    nx.draw(gr, node_size=50, pos=positions)
    plt.show()
```

P1.8 Draw square lattice with your own function. 

```python
def draw_square_lattice(n, m):
    no_nodes = n*m
    positions = {}
    am = np.zeros((no_nodes, no_nodes))
    i, j = 0, 0
    for node_id in range(no_nodes):
        positions[node_id] = (j, -i)
        if not (i == 0 and j == 0):
            if j != 0: 
                am[node_id, node_id-1] = 1
            if i != 0: 
                am[node_id, node_id-m] =1  
        j += 1
        if j == m: 
            j = 0 
            i += 1
    draw_graph_from_am(am, positions)
draw_square_lattice(10, 10)
```

P1.9 Draw chain with your own function. 

```python
def draw_chain(no_nodes):
    am = np.zeros((no_nodes, no_nodes))
    i, j = 0, 0
    for node_id in range(1, no_nodes):
        if node_id != 0: 
            am[node_id, node_id-1] = 1
    draw_graph_from_am(am)
draw_chain(20)
```

P1.10 Draw complete graph with your own function. 

```python
def draw_complete_graph(no_nodes):
    am = np.ones((no_nodes, no_nodes))
    am = np.tril(am, k=1)
    draw_graph_from_am(am, positions=[(math.cos(angle * 3.1415 / 180), math.sin(angle* 3.1415 / 180)) for angle in range(0, 360, 360 // no_nodes)])
draw_complete_graph(8)
```

P1.11 Draw square lattice with build-in function. 

```python
def draw_nx_square_lattice(n, m):
    G = nx.grid_2d_graph(n, m)
    positions = {n: n for n in G.nodes()}
    nx.draw(G, node_size=50, pos=positions)
    plt.show()
draw_nx_square_lattice(10, 10)
```

P1.12 Draw chain with build-in function. 

```python
def draw_nx_chain(n):
    G = nx.path_graph(n)
    nx.draw(G, node_size=50)
    plt.show()
draw_nx_chain(20)
```

P1.13 Draw complete graph with build-in function. 

```python
def draw_nx_complete_graph(n):
    G = nx.complete_graph(n)
    nx.draw(G, node_size=50, pos=[(math.cos(angle * 3.1415 / 180), math.sin(angle* 3.1415 / 180)) for angle in range(0, 360, 360 // n)])
    plt.show()
draw_nx_complete_graph(8)
```
